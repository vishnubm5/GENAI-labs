```python
from transformers import pipeline

sentiment = pipeline("sentiment-analysis")

texts = [
    "I loved the new movie! The acting was great.",
    "This product was terrible and broke in two days."
]

results = sentiment(texts)

for t, r in zip(texts, results):
    print(t)
    print(r)
    print()
```

```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer("all-MiniLM-L6-v2")

sentences = [
    "How do I cook pasta?",
    "Best way to boil spaghetti",
    "Capital of France?"
]

embeddings = model.encode(sentences, convert_to_tensor=True)

query = "How to make spaghetti?"
q_emb = model.encode(query, convert_to_tensor=True)

scores = util.pytorch_cos_sim(q_emb, embeddings)[0]

for s, score in zip(sentences, scores):
    print(round(score.item(), 3), s)
```

```python
from datasets import load_dataset

ds = load_dataset("imdb")
ds
```

```python
small_train = ds["train"].shuffle(seed=42).select(range(4000))
small_test = ds["test"].shuffle(seed=42).select(range(2000))
```

```python
import pandas as pd

train_df = pd.DataFrame(small_train)
test_df = pd.DataFrame(small_test)

train_df.head()
```

```python
from fastai.text.all import *

dls = TextDataLoaders.from_df(
    train_df,
    text_col="text",
    label_col="label",
    valid_pct=0.1,
    bs=32
)
```

```python
learn = text_classifier_learner(
    dls,
    AWD_LSTM,
    drop_mult=0.5,
    metrics=accuracy
)

from fastai.callback.progress import ProgressCallback
learn.remove_cbs(ProgressCallback)

learn.fine_tune(1)
```

```python
from datasets import load_dataset
from transformers import AutoTokenizer

dataset = load_dataset("glue", "sst2")
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")

def tokenize_fn(batch):
    return tokenizer(
        batch["sentence"],
        truncation=True,
        padding="max_length",
        max_length=128
    )

tokenized = dataset.map(tokenize_fn, batched=True)

tokenized["train"][0]
```

```python
import csv

texts = [
    "I love this phone.",
    "Worst purchase ever."
]

results = sentiment(texts)

with open("sentiment_results.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["text", "label", "score"])
    for t, r in zip(texts, results):
        writer.writerow([t, r["label"], r["score"]])

print("CSV saved")
```
